#  Decode Labs Internship — Project 1: Data Cleaning and Preparation

##  Project Objectives:
* *Identify* missing or null values and handle them effectively.
* *Remove* duplicate records to ensure data integrity.
* *Correct* data formats for dates, numbers, and text columns.

##  Step 1: Data Ingestion and Structural Audit
* Loading the Excel data source into a Pandas DataFrame to inspect total records, features, and the initial schema matrix.

```python
import pandas as pd

# Define data source path
file_path = "store_data2.xlsx"
df = pd.read_excel(file_path)

# Check data preview
df.head(2)
print("File Opened  Successfully!")

# Data shape check  (df.shape) 
print(f"Total Rows (Observations): {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
```

##  Step 2: Missing Data Profiling & Duplicate Check
* *Missing Data:* Identified null rows. Filled missing CouponCode entries with 'NO_COUPON' to keep data rows intact.
* *Duplicates:* Scanned the dataset for identical records and duplicate OrderID rows to ensure unique transaction tracking.

```python
#Check for missing values or null values in each column
print("Count of Missing (Null) values by Column:")
print("-" * 50)
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# Missing data percentage check
print("Percentage of Missing Values:")
print("-" * 50)
null_percentage = (df.isnull().sum() / len(df)) * 100
print(null_percentage)
```

```python
# Fill missing coupon code with 'NO_COUPON'
df["CouponCode"] = df["CouponCode"].fillna("NO_COUPON")
print("Missing values handled successfully!")

# Re-check if any missing value are left
print(df["CouponCode"].isnull().sum())
```

```python
# Find duplicate rows in the entire dataset
duplicate_rows_count = df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_rows_count}")

if duplicate_rows_count > 0:
    
    # Remove duplicate rows
    df = df.drop_duplicates()
    print("Duplicate rows removed successfully")
```

```python
# Duplicate OrderID

duplicate_ids_count = df.duplicated(subset=['OrderID']).sum()

print(f"Total number of duplicate OrderIDs found: {duplicate_ids_count}")

if duplicate_ids_count > 0:
    
    # Remove duplicate OrderIDs
    df = df.drop_duplicates(subset=['OrderID'])
    print("Duplicate OrderIDs removed successfully")
```

##  Step 3: Correcting Data Formats (Dates, Numbers, & Text)
* *Dates:* Converted the Date column to standard datetime format.
* *Numbers:* Fixed data types for Quantity, ItemsInCart, UnitPrice, and TotalPrice into clean integers and floats.
* *Text:* Stripped trailing spaces and converted categorical columns to UPPERCASE for uniformity.

```python
# Correct Date Format
# Convert date column to standard datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python

# Correct number formats
# Quantity and ItemsInCart to Integer format
df['Quantity'] = df['Quantity'].astype(int)
df['ItemsInCart'] = df['ItemsInCart'].astype(int)

# Convert UnitPrice and TotalPrice to Float format (decimal number) 
df['UnitPrice'] = df['UnitPrice'].astype(float)
df['TotalPrice'] = df['TotalPrice'].astype(float)
```

```python

# Correct text formats
# Convert all text-based columns to string format
text_columns = ['OrderID', 'CustomerID', 'Product', 'ShippingAddress', 'PaymentMethod', 'OrderStatus', 'TrackingNumber', 'CouponCode', 'ReferralSource']
for col in text_columns:
    # Convert to uppercase and remove leading/trailing spaces
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.upper()  
```

##  Step 4: Final Integrity Check & Data Export
* Verified all corrected data types and previewed the final rows to ensure zero errors.
* Exported the clean, production-ready dataset as a CSV file (Final_Cleaned_Ecom_Data.csv) for future analysis.

```python
# Final check for data types after corrections
print(" Update Data Format ")
print(df.dtypes)
```

```python
# Short summary of cleaned data types  first few rows
print("--- Cleaned Data Types ---")
print(df[['Date', 'Quantity', 'UnitPrice']].dtypes)

print("\n First 10 rows of cleaned data:")
display(df.head(10))
```

```python
# finally save the cleaned data to a new CSV file
df.to_csv("Final_Cleaned_Ecom_Data.csv", index=False)
print("'Final_Cleaned_Ecom_Data.csv' saved successfully")
```
